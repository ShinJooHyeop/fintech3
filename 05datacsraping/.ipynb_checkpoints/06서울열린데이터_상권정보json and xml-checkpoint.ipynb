# 서울시 상권 데이터(json)

```python
import requests
import pandas as pd
from dotenv import load_dotenv
import os
import re
load_dotenv()
import time
```

```python
url = f'http://openapi.seoul.go.kr:8088/{os.getenv("ss_api")}/json/VwsmAdstrdNcmCnsmpW/1/5/'
payload = dict(KEY=os.getenv("ss_api"),Type="json",pIndex=1, pSize=1000)
r = requests.get(url,params = payload)
print(r.url)
print(r.status_code)
responese = r.json()
responese
```

```python
total_pages = responese['VwsmAdstrdNcmCnsmpW']['list_total_count']//1000 +1
```

```python
responese['VwsmAdstrdNcmCnsmpW']['row']
```

```python
total_pages = responese['VwsmAdstrdNcmCnsmpW']['list_total_count']//1000 +1
total_pages
```

```python
result = {}
for item in responese['VwsmAdstrdNcmCnsmpW']['row']:
    for key, value in item.items():
        if type(value) == float:
            result.setdefault(key, []).append(int(value))
        else:
            result.setdefault(key, []).append(value)
df = pd.DataFrame(result)
df
```

```python
start = 1
end = 1000

result = {}
pages_num = 1
total_pages = 1
while pages_num <= total_pages:
    url = f'http://openapi.seoul.go.kr:8088/{os.getenv("ss_api")}/json/VwsmAdstrdNcmCnsmpW/{start}/{end}/'
    payload = dict(KEY=os.getenv("ss_api"),Type="json",pIndex=pages_num, pSize=1000)
    r = requests.get(url,params = payload)
    responese = r.json()
    
    if total_pages == 1:
        total_pages = responese['VwsmAdstrdNcmCnsmpW']['list_total_count']//1000 +1
    
    
    for item in responese['VwsmAdstrdNcmCnsmpW']['row']:
        for key, value in item.items():
            if type(value) == float:
                result.setdefault(key, []).append(int(value))
            else:
                result.setdefault(key, []).append(value)
    
    print(f"{pages_num}/{total_pages}현재 로딩중입니다",end = '\r')
    start += 1000
    end += 1000
    pages_num += 1
    
    time.sleep(0.2)
    
df = pd.DataFrame(result)
df
```

```python
df.to_csv("./data/json로 서울시 상권데이터 분석.csv",index = False , encoding = 'utf-8-sig')
```

# 서울시 상권 데이터(xml)

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
from dotenv import load_dotenv
import os
import re
load_dotenv()
import time
```

```python
url = f'http://openapi.seoul.go.kr:8088/{os.getenv("ss_api")}/xml/VwsmAdstrdNcmCnsmpW/1/5/'
payload = dict(KEY=os.getenv("ss_api"),Type="xml",pIndex=1, pSize=1000)
r = requests.get(url,params = payload)
response = bs(r.text, 'xml')
response
```

```python
total_pages = int(response.select_one('list_total_count').text) //1000 +1
total_pages
```

```python
tag_list = [tags.name for tags in response.select_one('row').find_all()]

print(type(int(response.select_one('list_total_count').text)))
```

```python
result2 = {}
for tag in tag_list:
        for value in response.select(tag):
            result2.setdefault(tag,[]).append(value.text)
df2 = pd.DataFrame(result2)
df2
```

```python
start = 1
end = 1000
page = 1
total_page = 1
result2 = {}

while page <= total_page:
    url =f'http://openapi.seoul.go.kr:8088/{os.getenv("ss_api")}/xml/VwsmAdstrdNcmCnsmpW/{start}/{end}/'
    r = requests.get(url)
    response = bs(r.text, 'xml')
    total_page = int(response.select_one("list_total_count").text) // 1000 + 1
    tag_list = [tags.name for tags in response.select_one('row').find_all()]

    kor_cols = ["기준_년분기_코드", "행정동_코드", "행정동_코드_명", "월_평균_소득_금액", "소득_구간_코드",
                "지출_총금액", "식료품_지출_총금액", "의류_신발_지출_총금액", "생활용품_지출_총금액",
                "의료비_지출_총금액", "교통_지출_총금액", "교육_지출_총금액", "유흥_지출_총금액",
                "여가_문화_지출_총금액", "기타_지출_총금액", "음식_지출_총금액"]
    
    for tag in tag_list:
        for value in response.select(tag):
            result2.setdefault(tag,[]).append(value.text)

    print(f"{page}/{total_page} 데이터 수집중", end="\r")            
    start += 1000
    end += 1000
    page += 1

    time.sleep(0.2)
        
df2 = pd.DataFrame(result2)
df2
```

```python
df2.to_csv("./data/xml로 서울시 상권데이터 분석.csv",index = False , encoding = 'utf-8-sig')
```
