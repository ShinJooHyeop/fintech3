```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn-v0_8')
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

df = pd.read_csv('./data/04_graph_practice_dataset.csv', encoding='cp949')
df['날짜'] = pd.to_datetime(df['날짜'])
df.head()
```

```python
df2 = df.copy()
```

```python
month_money= df[['매출','날짜']].groupby('날짜').sum()
month_money['매출']
```

```python
#문제1 월별 전체 매출 추이
plt.figure(figsize=(7,4))
plt.plot(month_money,marker='o',linewidth=2)
plt.title('월별 매출 추세')
plt.xlabel('날짜')
plt.ylabel('매출')
plt.grid(True, linestyle='--', alpha=0.4)
plt.show()
```

```python
#문제2 지역별 매출 추세 비교
plt.figure()
re_mo = df.groupby('지역')['매출'].sum()
re_mo.plot(kind='line')
plt.legend()
plt.show()
```

```python
#문제3 지역별 평균 매출
plt.figure(figsize=(6,4))
m = df.groupby('지역')['매출'].mean()
m.plot(kind='bar')
plt.title('지역별 평균 매출')
plt.xlabel('지역')
plt.ylabel('매출')
plt.show()
```

```python
#문제4 지역vs 제품별 매출
df_mean = df.groupby(['지역', '제품'])['매출'].mean().reset_index()
sns.barplot(data=df_mean, x='지역', y='매출', hue='제품')
plt.title('지역별 제품 평균 매출 비교')
```

```python
#문제5 제품별 총매출 비중
labels = ['A','B','C']
data = df.groupby('제품')['매출'].sum()
plt.figure(figsize=(5,5))
plt.pie(data, labels=labels, autopct='%.1f%%', startangle=90,
        wedgeprops={'edgecolor': 'white'})
plt.title('제품별 매출 비중')
plt.show()
```

```python
#문제6 이익 분포
plt.figure()
sns.histplot(df['이익'],bins=15,stat='density')
plt.title('이익 분포')
plt.show()
```

```python
#문제7 제품별 이익 분포 
plt.figure()
sns.boxplot(data = df2 ,x = '제품',y='이익',hue='제품')
plt.title('제품별  이익 분포')
plt.show()
```

```python
#문제8
plt.figure()
sns.scatterplot(data = df2, x = '매출',y = '이익',style='지역')
plt.title('매출과 이익 관계')
plt.show()
```

```python
#문제9 지역x제품 매출 Heatmap
df2['평균매출'] = df2.groupby('지역')['매출'].mean()
df2
```

```python
table = df2.pivot_table(index='지역', columns='제품', values='매출', aggfunc='mean')
plt.figure()
sns.heatmap(table, annot=True, fmt=".1f")
plt.title('지역-제품 쳥균 매출 히트맵')
```

```python
#문제10 매출 고객수
# 왼쪽 축: (막대)
months = ['1월','2월','3월','4월','5월','6월','7월','8월','9월','10월','11월','12월']
fig, ax1 = plt.subplots(figsize=(8,4))
ax1.bar(months, month_money['매출'], color='lightblue', alpha=0.7, label='지역x제품')
ax1.set_ylabel('매출(억원)')
ax1.set_xlabel('월')

# 오른쪽 축: (선)
ax2 = ax1.twinx()
ax2.plot(months, month_money['매출'], color='red', linewidth=2, label='평균매출')
ax2.set_ylabel('평균매출', color='red')

lines_1, labels_1 = ax1.get_legend_handles_labels()
lines_2, labels_2 = ax2.get_legend_handles_labels()

ax1.legend(lines_1 + lines_2, labels_1 + labels_2, loc='upper left', frameon=False)

plt.tight_layout()
plt.show()
```

```python
#문제11 로그 스케일 시각화
pro_mean = df2.groupby('제품')['고객수'].mean()

plt.figure()
pro_mean.plot(kind='bar')
plt.title('제품별 평균 고객 수 (일반 축)')
plt.ylabel('고객수')
plt.show()

plt.figure()
pro_mean.plot(kind='bar')
plt.title('제품별 평균 고객 수 (로그 축)')
plt.ylabel('고객수')
plt.yscale('log')
plt.show()
```

```python
#문제12 이상치 강조
monthly_sales = df.groupby('날짜')['매출'].sum()
mean_sales = monthly_sales.mean()

# --- 2. 조건부 색상 지정 ---
# 월별 매출이 평균보다 크면 'red', 아니면 'blue'
colors = ['red' if sales > mean_sales else 'blue' for sales in monthly_sales.values]

# --- 3. 막대 그래프 그리기 ---
plt.figure(figsize=(10, 6))

# x축: 월 (monthly_sales.index), y축: 매출 (monthly_sales.values)
plt.bar(monthly_sales.index, monthly_sales.values, color=colors)


# 그래프 제목 및 레이블 설정
plt.title('월별 매출 (평균 대비 강조)')
plt.xlabel('월')
plt.ylabel('총 매출')
plt.legend()
plt.show()
```

```python
#문제13
cum_sum = df2.groupby('날짜')['고객수'].sum()

plt.figure()
plt.plot(cum_sum, marker='o', linewidth=2)
plt.title('월별 고객 수 추이')
plt.xlabel('날짜')
plt.ylabel('고객수')
plt.grid(True, linestyle='--', alpha=0.4)

plt.annotate('고객 급증',
             xy=(cum_sum.index[6], 14385), xytext=(cum_sum.index[8], 14400),
             arrowprops=dict(arrowstyle='->', lw=1.5))
             
plt.show()
```
