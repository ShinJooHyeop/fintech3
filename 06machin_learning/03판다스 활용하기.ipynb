# concat으로 합치기

````python
# concat은 기본적으로 행방향으로 합치는 것
pd.concat([ticket,survived])
````

````python
# 열방 향으로 합칠라면 axis =1 
pd.concat([ticket,survived], axis=1)
````

# merge: SQL의 조인과 아주 유사 기본 inner

````python
# select * from passenger as t left join survived as s on t.passengerid = s.passengerid
pd.merge(passenger,survived,how='left',on='PassengerId')
````

````python
# select * from ticket as t left join survived as s on t.passengerid = s.passengerid
pd.merge(passenger,survived,how='inner',on="PassengerId")
````

````python
# select * from passenger as t left join survived as p on t.passengerid = s.passengerid
pd.merge(pd.merge(passenger,survived,how='right',on="PassengerId"),ticket, on='PassengerId')
````
