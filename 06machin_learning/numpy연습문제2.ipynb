```python
import pandas as pd
import numpy as np
```

```python
data = {'country': ['Belgium', 'France', 'Germany', 'Netherlands', 'United Kingdom'],
        'population': [11.3, 64.3, 81.3, 16.9, 64.9],
        'area': [30510, 671308, 357050, 41526, 244820],
        'capital': ['Brussels', 'Paris', 'Berlin', 'Amsterdam', 'London']}
countries = pd.DataFrame(data)
countries = countries.set_index('country') # 인덱스 지정
countries
```

```python
countries.info()
```

```python
#문제1
```

```python
density = (countries['population'] * 1000000)/countries['area']
```

```python
countries['density'] = density
```

```python
countries
```

```python
# 문제2 인구밀도가 300을 초과하는 국가(country)의 수도(capital)과 인구(population)을 선택해주세요.  
```

```python
countries[countries['density'] > 300]
```

```python
#문제3 density_ratio 칼럼을 추가해주세요. (density_ratio = 인구밀도/평균 인구밀도)
density_ratio = density/density.mean()
density_ratio
```
