````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import StratifiedKFold, validation_curve
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import make_scorer, f1_score
import koreanize_matplotlib
from utils import ohe_transform
````

````python
data = pd.read_csv("./data/salary2.csv")
data.head()
````

````python
data = data.drop_duplicates()
data = data.dropna()
data = data.reset_index(drop=True)
data['class'] = data['class'].apply(lambda x: 1 if x == ' >50K' else 0)
data['class'].value_counts()
X = data.drop('class', axis=1)
y = data['class']
````

````python
from sklearn.model_selection import train_test_split
````

````python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, stratify=y, random_state=10)
````

````python
X_train, X_test = ohe_transform(X_train, X_test)
````

````python
from sklearn.model_selection import learning_curve
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import (
    make_scorer, f1_score,
    precision_recall_curve, average_precision_score, auc,
    roc_curve, roc_auc_score)
````

````python
param_range = np.arange(1, 21)  # 1~20
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
````

````python
# κ²€μ¦κ³΅μ„ : max_depth λ³€ν™” vs F1 μ μ
train_scores, val_scores = validation_curve(
    DecisionTreeClassifier(random_state=42, class_weight='balanced'),
    X_train, y_train, param_name="max_depth", param_range=param_range,
    cv=cv, scoring=make_scorer(f1_score)
)

# ν‰κ·  μ μ κ³„μ‚°
train_mean = train_scores.mean(axis=1)
val_mean = val_scores.mean(axis=1)

# λ‹¨μΌ ν”λ΅― (μƒ‰μƒ μ§€μ • μ—†μ)
plt.figure(figsize=(6,4))
plt.plot(param_range, train_mean, marker="o", label="ν›λ ¨ F1")
plt.plot(param_range, val_mean, marker="s", label="κµμ°¨κ²€μ¦ F1")
plt.title("κ²€μ¦κ³΅μ„  (DecisionTree: max_depth)")
plt.xlabel("max_depth")
plt.ylabel("F1 μ μ")
plt.legend()
plt.show()
````

> λ”°λΌμ„ **ν›λ ¨/κ²€μ¦ μ„±λ¥μ΄ κ· ν• μ΅ν μ§€μ **μ΄ κ°€μ¥ μΆ‹μ€ `max_depth`μ΄λ‹¤.



## 5) ν•™μµκ³΅μ„ (Learning Curve)μΌλ΅ λ°μ΄ν„°/λ¨λΈ ν•κ³„ νμ•…

ν›λ ¨ λ°μ΄ν„° μ–‘μ„ λλ ¤κ°€λ©° **ν›λ ¨/κ²€μ¦ μ„±λ¥**μ΄ μ–΄λ–»κ² λ³€ν•λ”μ§€ λ³΄λ©΄,

- λ°μ΄ν„°κ°€ 7000κ° μ΄μƒλ¶€ν„° ν›λ ¨ λ°μ΄ν„°κ°€ λ§λ‹¤κ³  ν•΄μ„ μ„±λ¥μ΄ μΆ‹μ•„μ§€λ”κ±΄ μ•„λ‹λ‹¤


````python
train_sizes, train_scores, val_scores = learning_curve(
    DecisionTreeClassifier(random_state=42, class_weight='balanced'),
    X_train, y_train, cv=cv, scoring=make_scorer(f1_score),
    train_sizes=np.linspace(0.2, 1.0, 5), shuffle=True, random_state=42
)

plt.figure(figsize=(6,4))
plt.plot(train_sizes, train_scores.mean(axis=1), marker="o", label="ν›λ ¨ F1")
plt.plot(train_sizes, val_scores.mean(axis=1), marker="s", label="κµμ°¨κ²€μ¦ F1")
plt.title("ν•™μµκ³΅μ„  (DecisionTree)")
plt.xlabel("ν›λ ¨ λ°μ΄ν„° ν¬κΈ°")
plt.ylabel("F1 μ μ")
plt.legend()
plt.show()
````

| **κ²°λ΅ ** | λ°μ΄ν„°μ ν¬κΈ°λ³΄λ‹¤ λ¨λΈμ λ³µμ΅λ„ μ μ–΄(`max_depth`, `min_samples_leaf`)κ°€ ν•„μ” |

| λ°μ΄ν„° λ¶κ· ν• | `class_weight='balanced'` μµμ… μ‚¬μ© |
| λ‹¨μΌ νΈλ¦¬μ ν•κ³„ | λ°°κΉ…(RandomForest), λ¶€μ¤ν…(LightGBM, XGBoost) ν™μ© |

## π“ 5. μ”μ•½

> ν›λ ¨ λ°μ΄ν„° ν¬κΈ°λ¥Ό λλ ¤λ„ κµμ°¨κ²€μ¦ μ„±λ¥μ΄ ν–¥μƒλμ§€ μ•λ” κ²ƒμ€  
> λ¨λΈμ΄ **λ°μ΄ν„° λ³µμ΅λ„μ— λΉ„ν•΄ μ§€λ‚μΉκ² λ³µμ΅**ν•κΈ° λ•λ¬Έμ΄λ‹¤.  
> μ¦‰, **DecisionTreeλ” ν„μ¬ κ³Όμ ν•© μƒνƒ**μ΄λ©°,  
> **ν•μ΄νΌνλΌλ―Έν„° νλ‹ λλ” μ•™μƒλΈ” κΈ°λ²• μ μ©**μ΄ ν•„μ”ν•λ‹¤.



## 6) μ–΄λ–¤ μ§€ν‘λ΅ νλ‹ν• κΉ?

- λ¶κ· ν• λ°μ΄ν„°μ΄λ©΄ **F1, Recall, PR-AUC**λ¥Ό κ¶μ¥ν•λ‹¤.  
- λΉ„μ© λ―Όκ° λ¬Έμ λΌλ©΄, μ—…λ¬΄ λΉ„μ©μ— λ§μ¶° **μ„κ³„κ°’**μ„ λ³„λ„λ΅ μ΅°μ •ν•λ‹¤.  
- ROC-AUCλ§ λ†’κ³  PR-AUCκ°€ λ‚®λ‹¤λ©΄, **ν¬μ† μ–‘μ„± λ¬Έμ **μΌ κ°€λ¥μ„±μ΄ ν¬λ‹¤.


````python
# =========================
# 2) Hold-outμ© PR λ³€μλ“¤ κ³„μ‚° (λ°λ“μ‹ ROC/PR μ „μ— μ •μ)
# =========================
dt = DecisionTreeClassifier(random_state=42, class_weight='balanced')
dt.fit(X_train, y_train)

# μμΈ΅ ν™•λ¥ (μ–‘μ„± ν΄λμ¤)
y_score = dt.predict_proba(X_test)[:, 1]

# PR-curve, PR-AUC(AP), κΈ°μ¤€μ„ 
precision, recall, thresholds = precision_recall_curve(y_test, y_score)
pr_auc = auc(recall, precision)
ap = average_precision_score(y_test, y_score)
baseline = (y_test == 1).mean()

# μµλ€ F1 μ§€μ 
f1_vals = 2 * precision * recall / (precision + recall + 1e-12)
best_idx = int(np.nanargmax(f1_vals))
best_recall = recall[best_idx]
best_precision = precision[best_idx]
best_f1 = f1_vals[best_idx]
best_threshold = thresholds[max(0, best_idx - 1)]  # thresholds κΈΈμ΄ λ³΄μ •

# =========================
# 3) ROC-AUC vs PR-AUC λΉ„κµ (Hold-out + μ„ νƒ: CV) - μ•μ „ν• ν¬λ§·ν…/ν”λ΅ν…
# =========================
from sklearn.metrics import roc_curve, roc_auc_score

def fmt3(x):
    return f"{x:.3f}" if x is not None else "N/A"

# --- Hold-out ROC ---
roc_auc = roc_auc_score(y_test, y_score)
fpr, tpr, _ = roc_curve(y_test, y_score)

print("=== Hold-out μ§€ν‘ μ”μ•½ ===")
print(f"ROC-AUC = {roc_auc:.3f} | PR-AUC = {pr_auc:.3f} | AP = {ap:.3f} | μ–‘μ„±λΉ„μ¨(κΈ°μ¤€μ„ ) = {baseline:.3f}")

# (μ„ νƒ) CV κΈ°λ° ROC/PRλ„ y_score_cvκ°€ μμ„ λ•λ§ κ³„μ‚°
roc_auc_cv = None
try:
    if 'y_score_cv' in locals():
        roc_auc_cv = roc_auc_score(y_train, y_score_cv)
        fpr_cv, tpr_cv, _ = roc_curve(y_train, y_score_cv)
        # pr_auc_cv, ap_cv λ” μ•μ try λΈ”λ΅μ—μ„ μ΄λ―Έ κ³„μ‚°λμ„ μλ„ μκ³  μ—†μ„ μλ„ μμ
except Exception as e:
    print(f"[INFO] CV ROC-curveλ” κ±΄λ„λ€: {e}")

# --- 2μ—΄ λΉ„κµ κ·Έλ¦Ό: (μΆ) ROC, (μ°) PR ---
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# (μΆ) ROC Curve
axes[0].plot(fpr, tpr, label=f"Hold-out ROC (AUC={roc_auc:.3f})")
axes[0].plot([0, 1], [0, 1], linestyle="--", label="λ¬΄μ‘μ„ κΈ°μ¤€μ„  (AUC=0.5)")
if (roc_auc_cv is not None) and ('fpr_cv' in locals()):
    axes[0].plot(fpr_cv, tpr_cv, label=f"CV ROC (AUC={fmt3(roc_auc_cv)})")
axes[0].set_xlabel("FPR (κ±°μ§“ μ–‘μ„± λΉ„μ¨)")
axes[0].set_ylabel("TPR (μ¬ν„μ¨)")
axes[0].set_title("ROC Curve")
axes[0].legend(loc="lower right")

# (μ°) Precisionβ€“Recall Curve
axes[1].step(recall, precision, where="post",
             label=f"Hold-out PR (AUC={pr_auc:.3f}, AP={ap:.3f})")
axes[1].hlines(baseline, 0, 1, linestyles="--",
               label=f"μ–‘μ„±λΉ„μ¨ κΈ°μ¤€μ„ ={baseline:.3f}")

# CV PRκ³΅μ„ : precision_cv/recall_cvκ°€ μκ³ , pr_auc_cv/ap_cvκ°€ κ³„μ‚°λ κ²½μ°λ§
if ('precision_cv' in locals()) and ('recall_cv' in locals()):
    label_cv = f"CV PR (AUC={fmt3(pr_auc_cv)}, AP={fmt3(ap_cv)})"
    axes[1].step(recall_cv, precision_cv, where="post", label=label_cv)

axes[1].scatter([best_recall], [best_precision], marker="o", zorder=5,
                label=f"μµλ€ F1 μ§€μ  (F1={best_f1:.3f}, thr={best_threshold:.3f})")
axes[1].set_xlim([0, 1])
axes[1].set_ylim([0, 1.05])
axes[1].set_xlabel("Recall (μ¬ν„μ¨)")
axes[1].set_ylabel("Precision (μ •λ°€λ„)")
axes[1].set_title("Precisionβ€“Recall Curve")
axes[1].legend(loc="lower left")

plt.tight_layout()
plt.show()
````

## π“ 3. Precision-Recall Curve (μ¤λ¥Έμ½ κ·Έλν”„)

- **xμ¶•:** Recall (μ¬ν„μ¨, μ‹¤μ  μ–‘μ„± μ¤‘ λ§μ¶ λΉ„μ¨)  
- **yμ¶•:** Precision (μ •λ°€λ„, μμΈ΅ν• μ–‘μ„± μ¤‘ μ‹¤μ  μ–‘μ„± λΉ„μ¨)

### μ£Όμ” νΉμ§•:
